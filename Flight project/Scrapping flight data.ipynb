```python
airline_name=[]
flightsDate=[]
departureTime=[]
arrivalTime=[]
Source=[]
Destination=[]
Duration=[]
Stops=[]
Price=[]

c=0
for i in links[::2]:
    driver.get(i)
    x=input("Got Link? (Y/N)")
    c+=1
    print("Link counter:",c)
    for j in range(9):
        print("\t\tStarted...")
        airline_name.extend([flightName.text for flightName in driver.find_elements_by_xpath("//span[@class='i-b text ellipsis']")])
        departureTime.extend([tim.text for tim in driver.find_elements_by_xpath("//div[@class='i-b pr']")])
        arrivalTime.extend([tim.text[:5] for tim in driver.find_elements_by_xpath("//p[@class='bold fs-15 mb-2 pr time']")])
        Source.extend([tim.text for tim in driver.find_elements_by_xpath("//p[@class='fs-10 font-lightgrey no-wrap city ellipsis']")[::2]])
        Destination.extend([tim.text for tim in driver.find_elements_by_xpath("//p[@class='fs-10 font-lightgrey no-wrap city ellipsis']")[1::2]])  
        Duration.extend([tim.text for tim in driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']")])
        Stops.extend([tim.text.split(" ")[0] for tim in driver.find_elements_by_xpath("//div[@class='stop-cont pl-13']/div/span")])
        Price.extend([tim.text.replace(",","") for tim in driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']")])
        date=driver.find_element_by_xpath("//div[@class='day-li text-center cursor-pointer pr active font-primary-color']").text
        for k in range(len(driver.find_elements_by_xpath("//div[@class='tuple']"))):
            flightsDate.append(date)
        
        print("\tChange Date counter:",j)
        if j in [2,5]:
            time.sleep(10)
        elif j==8:
            continue
        else:
            time.sleep(3)
        
```
